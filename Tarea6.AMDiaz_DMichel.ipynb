```python
# Código de solución estudiante 2

# Función de aceptación y rechazo usando compresión de listas (vista en clase 10)
def Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
#     R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'oc') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
     for i in range(len(R1))] 
    plt.show()
    
    x = [R1[i] for i in range(len(R1)) if condition[i]]
    
    return x

# Parámetros de la función 
N = 500  # número de puntos
x =np.arange(0,1,0.01)
# Función objetivo 
f = lambda x: 30*(x**2 -2 * x**3 + x**4) 
# Máximo de la función f
a = f(optimize.fmin(lambda x:-f(x), 0, disp=False))
plt.plot(x,f(x))

t = lambda x: a * np.sin(np.pi*x)
plt.plot(x,f(x),'m',lw=5,label='f(x)')
plt.plot(x,t(x),'g',lw=5,label='t(x)')
plt.legend()
plt.title("Gráfica",fontsize=14)
#plt.xlabel('Rango',fontsize=14)
#plt.ylabel('Densidad de probabilidad',fontsize=14)
R1 = np.random.rand(N)
R2 = np.random.rand(N)
Acep_rechazo(R2,R1,f,t)
plt.show()
```

si:$$P(k+1) = \frac{\lambda}{K+1}P(k)$$Entonces:$$\frac{P(k+1)}{P(k)} = \frac{\lambda}{k+1}$$Usando la fórmula recursiva para calcular $P(k+1)$ en términos de $P(k)$:$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$También se hace uso de la propiedad$$\frac{k!}{(k+1)!}=\frac{1}{k+1}$$$$\frac{P(k+1)}{P(k)} = \frac{\lambda^{k+1}k!}{\lambda^{k}(k+1)!} = \frac{\lambda}{k+1}$$Por lo que:$$P(k+1) = \frac{\lambda}{k+1}P(k)$$


```python
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       indices: 'valores reales a generar aleatoriamente',
                       N: 'cantidad de números aleatorios a generar'):
    
    U =np.random.rand(N)
    # Diccionario de valores aleatorios
    rand2reales = {i: idx for i, idx in enumerate(indices)}

    # Series de los valores aletorios
    y = pd.Series([sum([1 for p in p_acum if p < ui]) for ui in U]).map(rand2reales)

    return y
def plot_histogram_discrete(distribucion:'señal de varibles aleatorias de un distribución DISCRETA dada',
                            label:'label del legend a aparecer en el gráfica',
                            densidad:'por defecto regresa el histograma en densidad'=True):
    # len(set(distribucion)) cuenta la cantidad de elementos distintos de la variable 'distribucion'
    plt.figure(figsize=[8,4])
    y, x = np.histogram(distribucion, bins=len(set(distribucion)), density=densidad) 
    plt.bar(x[1:], y, label=label)
    plt.legend()
    plt.grid()
    plt.title('Poisson',fontsize=14)
    plt.xlabel('Rango',fontsize=14)
    plt.ylabel('Probabilidad',fontsize=14)
    plt.show()  
# Función para la distribución acumulada de Poisson
#hecha como la P_acum_Binomial_o de la clase 10:
def poisson(l,k):
    z=np.zeros(k+1)
    z[0]=np.exp(-l)
    def z1(i):
        nonlocal z
        z[i+1]=z[i]*(l/(i+1))
    # Lleno el vector Pr usando compresión de listas
    [z1(i) for i in range(k-1)]
    return np.cumsum(z)
N = 10000;k = 10
Poisson=poisson(5,5)
# Usando el método de la transformada inversa
pl= Gen_distr_discreta(Poisson,np.arange(0,k+1),N)    
plot_histogram_discrete(pl,"Distribución Poisson")
```
